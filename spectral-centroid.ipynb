```python
import matplotlib.pyplot as plt 
import numpy as np 
import librosa
```

```python
debussy_file = "audio/debussy.wav"
redhot_file = "audio/redhot.wav"
duke_file = "audio/duke.wav"
```

```python
# load audio files with librosa
debussy, sr = librosa.load(debussy_file)
redhot, _ = librosa.load(redhot_file)
duke, _ = librosa.load(duke_file)
```

```python
FRAME_SIZE = 2048
HOP_SIZE = 512
```

```python
sc_debussy = librosa.feature.spectral_centroid(debussy, sr=sr, hop_length=HOP_SIZE, n_fft=FRAME_SIZE)[0]
sc_duke = librosa.feature.spectral_centroid(duke, sr=sr, hop_length=HOP_SIZE, n_fft=FRAME_SIZE)[0]
sc_redhot = librosa.feature.spectral_centroid(redhot, sr=sr, hop_length=HOP_SIZE, n_fft=FRAME_SIZE)[0]
```

```python
sc_debussy.shape
```

```python
frames = range(len(sc_debussy))
t = librosa.frames_to_time(frames, sr=sr, hop_length=HOP_SIZE, n_fft=FRAME_SIZE)
```

```python
# Visualise Spectral Centroid

plt.figure(figsize=[20,10])

plt.plot(t, sc_debussy, color='b')
plt.plot(t, sc_redhot, color='r')
plt.plot(t, sc_duke, color='y')
```
